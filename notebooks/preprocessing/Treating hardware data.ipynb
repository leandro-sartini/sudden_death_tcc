```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('T0000.CSV',sep=',',encoding = 'UTF-8',skiprows=15)
```

```python
df
```

```python
plt.figure(figsize = [16,8])
sns.lineplot(data=df,x=df.index,y='CH1')
```

```python
df['smoothed'] = pd.Series(df['CH1']).rolling(window=1000).mean()
```

```python
plt.figure(figsize = [16,8])
line_plot = sns.lineplot(data=df,x=df.index,y='smoothed')
fig = line_plot.get_figure()
fig.savefig('test.png')
```
